```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MonteCarlo( price, time, riskFreeRate, beta, steps, N):
    dt= time/steps
    ST= np.log(price)+ np.cumsum(((riskFreeRate- beta** 2/2 )* dt + beta * np.sqrt(dt) * np.random.normal(size=( steps, N))), axis=0)
    print(np.exp(ST).mean())
    return np.exp(ST)
```

```python

paths=MonteCarlo(171.35, 1, .03, 1.33, 242, 1000)
plt.plot(paths)
plt.xlabel("Time in days")
plt.ylabel("Price")
plt.title("Geometric Brownian Motion META")
```

```python

paths=MonteCarlo(174.31, 1, .03, 0.65, 242, 60)
plt.plot(paths)
plt.xlabel("Time increment in days")
plt.ylabel("Stock Price")
plt.title("NVIDIA")
```
